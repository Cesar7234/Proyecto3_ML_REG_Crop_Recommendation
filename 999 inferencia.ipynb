```python
# model_inference.py
import joblib
import pandas as pd
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Cargar el modelo y el escalador
model = joblib.load('decision_tree_model.pkl')
scaler = joblib.load('scaler.pkl')

# Cargar los datos de prueba
X_train, X_test, y_train, y_test = joblib.load('preprocessed_data.pkl')
#df_test = pd.read_csv('new_data.csv')  # O usa los datos de prueba originales
X_new = X_test

# Realizar predicciones
y_pred = model.predict(X_new)

# Comparar predicciones con valores reales (si los tienes)
y_real = y_test
mse = mean_squared_error(y_real, y_pred)
print(f'MSE en datos nuevos: {mse}')

# Visualizar resultados
plt.scatter(y_real, y_pred)
plt.xlabel('Valores reales')
plt.ylabel('Predicciones')
plt.title('Predicciones vs Valores reales')
plt.show()
```
